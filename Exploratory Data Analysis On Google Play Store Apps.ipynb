### Analyze Size column

```python
df['Size'].dtype
```

```python
df['Size'].nunique()
```

```python
df['Size'].unique()
```

```python
len(df[df['Size']=='Varies with device'])
```

#### There are 1227 Apps on Play Store which has size varying with device. These Apps are compatable for all devices.

#### Convert each app size in KB first then MB(I Iknow 1MB=1024KB, but for simplisity I am taking 1MB=1000KB)
#### Replacing "varies with device"  to median of Size column.

```python
df['Size']=df['Size'].apply(lambda x : x.replace('M', '000').replace('k', ''))
df['Size']=df['Size'].replace('Varies with device', np.nan)
df['Size']=df['Size'].fillna(df['Size'].median())
df['Size']=df['Size'].astype(float)
df['Size(MB)']=(df['Size']/1000).round(3)
df.drop('Size', axis=1, inplace=True)
df['Size(MB)'].unique()
```

#### Largest, Lowest and average App size

```python
print("Maximum Size",df['Size(MB)'].max())
print("Minimum Size",df['Size(MB)'].min())
print("Average Size",df['Size(MB)'].mean())
```
