IQR

```python
q1=df['Income'].quantile(0.25)
q3=df['Income'].quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
IQR=q3-q1
```

```python
IQR
```

```python
lower_bound=q1 - 1.5*IQR
upper_bound=q3 + 1.5*IQR
```

```python
lower_bound
```

```python
upper_bound
```

```python
cleaned_df=df[(df['Income']<upper_bound) & (df['Income']>lower_bound)]
```

```python
cleaned_df['Income'].plot(kind='kde')
```

```python
cleaned_df['Income'].plot(kind='hist')
```

```python
cleaned_df['Income'].plot(kind='box')
```
